# Operational KPI Calculator

This notebook calculates operational efficiency for three sites, classifies their operational status and prints a formatted report.

## Calculate Efficiency

Efficiency measures how well a site performs compared to its target.
Formula:

Efficiency (%) = (Actual Output / Target Output) × 100

We create a function so that the same calculation can be reused for multiple sites.

```python
def calculate_efficiency(actual_output, target_output):
    """
    Calculate operational efficiency as a percentage.
    """
    return (actual_output / target_output) * 100
```

## Determine Operational Status

After calculating efficiency, we classify the site's performance.
Rules:

- Critical: Efficiency is below 70%
- Warning: Efficiency is between 70% and 89%
- Normal: Efficiency is 90% or higher

This helps management quickly identify which sites need attention.

```python
def get_operational_status(efficiency):
    """
    Classify operational performance based on efficiency.
    """
    if efficiency < 70:
        return "Critical"
    elif 70 <= efficiency <= 89:
        return "Warning"
    else:
        return "Normal"
```

## Create Operational Site Data

We create a dictionary containing information for three fuel depots.

Each depot contains:

- name
- actual_output
- target_output

This structure makes it easy to store and process operational information.

```python
sites = {
    "site1": {
        "name": "Nairobi Fuel Depot",
        "actual_output": 6500,
        "target_output": 10000
    },
    "site2": {
        "name": "Mombasa Fuel Depot",
        "actual_output": 8200,
        "target_output": 10000
    },
    "site3": {
        "name": "Kisumu Fuel Depot",
        "actual_output": 9500,
        "target_output": 10000
    }
}
```

## Generate Operational Reports

We loop through each site and:

1. Calculate efficiency.
2. Determine operational status.
3. Print a formatted report.

This approach is scalable because new sites can be added without changing the analysis logic.

```python
print("OPERATIONAL PERFORMANCE REPORT")
print("-" * 40)

for site in sites.values():
    
    efficiency = calculate_efficiency(
        site["actual_output"],
        site["target_output"]
    )
    
    status = get_operational_status(efficiency)
    
    print(f"Site Name      : {site['name']}")
    print(f"Actual Output  : {site['actual_output']}")
    print(f"Target Output  : {site['target_output']}")
    print(f"Efficiency (%) : {efficiency:.2f}%")
    print(f"Status         : {status}")
    print("-" * 40)
```
